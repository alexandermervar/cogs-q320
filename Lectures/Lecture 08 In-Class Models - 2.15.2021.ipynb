```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
investment = 100
starting = 5
end = 65
duration = end - starting
interest = 0.07
fee = 0
feesTotal = 0
annualInvestment = 1200
```

```python
# Equation = s[t] = s[t-1] + s[t-1]*i + c
```

```python
moneyOverTime = np.zeros(duration)
feesOverTime = np.zeros(duration)
for i in range(duration):
    moneyOverTime[i] = investment
    feesOverTime[i] = feesTotal
    investment += investment*interest + annualInvestment
    feesTotal += investment*fee
    investment -= investment*fee
    
print("Investment: " + str(investment))
```

```python
plt.plot(moneyOverTime)
plt.xlabel("Years over Time")
plt.ylabel("Money in Account")
plt.show()
```

```python
plt.plot(feesOverTime)
plt.xlabel("Years over Time")
plt.ylabel("Money Paid in Fees")
plt.show()
```
